```python
import matplotlib.pyplot as plt
import yfinance as yf
import pandas as pd
import numpy as np

# Define the ticker list
tickers = ['AAPL', 'VZ', 'KO', 'PEP']

# Fetch the data
data = yf.download(tickers,'2018-01-01')
daily_close = data['Adj Close']
                   
# Plot the adusted closing price
data['Adj Close'].plot()
plt.ylabel('Price')
plt.xlabel('Date')

# Daily return
daily_pct_change = daily_close.pct_change()

# Plot the adjusted closing prices
cum_daily_return = (1 + (daily_pct_change)).cumprod()
cum_daily_return.plot()
plt.ylabel('Price')
plt.xlabel('Date')

# Volatility
min_period = 60 
volatility = daily_pct_change.rolling(min_period).std() * np.sqrt(min_period) 
volatility.plot()
```
