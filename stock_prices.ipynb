```python
import yfinance as yf
import pandas as pd

# Define the ticker list
tickers_list = ['AAPL', 'VZ', 'MSFT', 'TSLA', 'INTC']
data = pd.DataFrame(columns=tickers_list)
# Fetch the data
for ticker in tickers_list:
    data[ticker] = yf.download(ticker, '2018-01-01','2020-01-01')['Adj Close']
# Print first 5 rows of the data
print(data.head())
# Plot all the close prices
((data.pct_change()+1).cumprod()).plot()
plt.ylabel('Price')
plt.xlabel('Date')
```

```python
import yfinance as yf
import pandas as pd
import pyfolio as pf

# Define the ticker list
tickers_list = ['AAPL', 'VZ', 'MSFT', 'TSLA', 'INTC']
data = pd.DataFrame(columns=tickers_list)
# Fetch the data
for ticker in tickers_list:
    data[ticker] = yf.download(ticker, '2018-01-01','2020-01-01')['Adj Close']

# Compute the returns of individual stocks and then compute the daily mean returns.
data = data.pct_change().dropna().mean(axis=1)
# Print first 5 rows of the data
print(data.head())
pf.create_full_tear_sheet(data)
```
